```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
a, b = 2.31, 0.627
```

```python
Beta = sts.beta(a,b)
```

```python
sample_beta = Beta.rvs(1000)
```

```python
x = np.linspace(0,1,1000)
```

```python
pdf_beta = Beta.pdf(x)
```

```python
plt.hist(sample_beta, normed = True)
plt.plot(x, pdf_beta)
plt.ylabel('$Beta(x, %s, %s)$' % (a,b))
plt.xlabel('$x$')
```

```python
def fun(n):
    sample_average = np.array(range(0,1000), float)
    for k in range(0,1000):
        sample_beta = Beta.rvs(n)
        sample_average[k] = sum(sample_beta)/n
    norm_beta = sts.norm(loc = a/(a+b), scale = math.sqrt(a*b/(a+b)**2/(a+b+1)/n))
    x = np.linspace(0,1.5,1000)
    pdf_norm = norm_beta.pdf(x)
    plt.hist(sample_average, normed=True)
    plt.plot(x, pdf_norm, 'ro')
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
```

```python
fun(5)
```

```python
fun(10)
```

```python
fun(50)
```
